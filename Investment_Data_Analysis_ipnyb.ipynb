## **TASK 01: Data Overview**

**Objective: Understanding the dataset structure.**

```python
import pandas as pd

# Read the Excel file into a pandas DataFrame
investment_data = pd.read_excel('/content/Investment Dataset.xlsx')

investment_data.info()
```

## **Task 02: Gender Distribution**

**Objective: Extracting and Visualize gender distribution in
the dataset.**

```python
# Assuming 'Gender' is the column name for gender information.
gender_column = investment_data['gender']
gender_column.head()
```

```python
import matplotlib.pyplot as plt
# Assuming 'Gender' is the column name for gender information
gender_counts = investment_data['gender'].value_counts()

# Pie chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')  # Remove the y-axis label for pie charts
plt.show()
```

### **Task 4: Most Preferred Investment Avenue**

**Identify the most preferred investment avenue**

```python
# Assuming 'Investment Avenue' is the column name for investment information.
investment_avenue_column = investment_data['Avenue']

# Count the occurrences of each investment avenue
investment_counts = investment_avenue_column.value_counts()

# Display the counts
print("\nInvestment Avenue Counts:")
print(investment_counts)

# Bar chart of investment avenues
plt.figure(figsize=(10, 6))
investment_counts.plot(kind='bar')
plt.title('Distribution of Investment Avenues')
plt.xlabel('Investment Avenue')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

# The investment avenue with the highest frequency
most_frequent_investment = investment_counts.idxmax()
print(f"\nThe most frequent investment avenue is: {most_frequent_investment}")
```

### **Task 10: Correlation Analysis**

**Objective: Establish the most correlated factors driving Investment**

```python
# Select relevant columns for correlation analysis
# Treating 'Duration' and 'Expect' as categorical, we'll use a different approach than Pearson correlation
selected_columns = investment_data[['age', 'Duration', 'Expect']]

# Convert 'Duration' and 'Expect' to categorical data type if not already
selected_columns['Duration'] = selected_columns['Duration'].astype('category')
selected_columns['Expect'] = selected_columns['Expect'].astype('category')

# Analyze relationships between 'age' and categorical variables using groupby and describe
print("\nRelationship between Age and Duration:")
print(selected_columns.groupby('Duration')['age'].describe())

print("\nRelationship between Age and Expected Return:")
print(selected_columns.groupby('Expect')['age'].describe())

# Visualize the relationship between age and other factors (categorical) using box plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
selected_columns.boxplot(column='age', by='Duration', rot=45)  # Rotate x-axis labels
plt.title('Age vs. Duration')
plt.xlabel('Duration')  # Add x-axis label
plt.ylabel('Age')  # Add y-axis label

plt.subplot(1, 2, 2)
selected_columns.boxplot(column='age', by='Expect', rot=45)  # Rotate x-axis labels
plt.title('Age vs. Expected Return')
plt.xlabel('Expected Return') # Add x-axis label
plt.ylabel('Age') # Add y-axis label

plt.tight_layout()
plt.show()

```
